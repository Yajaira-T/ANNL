Esto es una prueba:
\begin{equation}
    \int_a^b f'(x)fx = f(b) - f(a)
\end{equation}

```python
1+1
```

```python
def punto_medio(f, a, b, N):
    h = (b - a) / N
    return h * sum(f(a + (i + 0.5) * h) for i in range(N))

def rectangulo(f, a, b, N):
    h = (b - a) / N
    return h * sum(f(a + i * h) for i in range(N))

def trapecio(f, a, b, N):
    h = (b - a) / N
    return h * ((f(a) + f(b)) / 2 + sum(f(a + i * h) for i in range(1, N)))
```

```python
f = lambda x: x**5

N= 50
print('Resultados para la regla del punto medio', punto_medio(f, 0, 10, N))
print('Resultados para la regla del rectangulo', rectangulo(f, 0, 10, N))
print('Resultados para la regla del trapecio',trapecio(f, 0, 10, N))
```
